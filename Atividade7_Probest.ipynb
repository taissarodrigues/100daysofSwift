```python
import pandas as pd

## Produtos da minha cesta e seus dados
dados = {
    "Produto": ["Arroz", "Feijão - mulatinho", "Feijão - preto", "Farinha de arroz",
                "Macarrão", "Massa semipreparada", "Macarrão instantâneo",
                "Batata-inglesa", "Balas"],
    "Código": ["1101002", "1101051", "1101052", "1102001", "1102006",
               "1102029", "1102061", "1103003", "1104018"],
    "Variação Nov/2024 (%)": [1.2, -2.5, 3.1, 0.8, -0.5, 0.3, -1.2, 2.3, 0.1],
    "Variação Dez/2024 (%)": [-1.8, 4.4, -3.9, 0.7, -0.2, 0.0, -0.4, -2.2, -0.1]  # Exemplo
}

pesos_medios = {
    "Arroz": 0.22,
    "Feijão - mulatinho": 0.15,
    "Feijão - preto": 0.10,
    "Farinha de arroz": 0.05,
    "Macarrão": 0.12,
    "Massa semipreparada": 0.08,
    "Macarrão instantâneo": 0.07,
    "Batata-inglesa": 0.15,
    "Balas": 0.06
}

df = pd.DataFrame(dados)
df["Peso Médio"] = df["Produto"].map(pesos_medios)

df["Contrib Nov"] = df["Variação Nov/2024 (%)"] * df["Peso Médio"] / 100
inflacao_nov = df["Contrib Nov"].sum()

df["Contrib Dez"] = df["Variação Dez/2024 (%)"] * df["Peso Médio"] / 100
inflacao_dez = df["Contrib Dez"].sum()

## Resultado
print("Resultados:")
print(f"Inflação da cesta em Nov/2024: {inflacao_nov:.4f}%")
print(f"Inflação da cesta em Dez/2024: {inflacao_dez:.4f}%")
print(f"Variação entre os meses: {inflacao_dez - inflacao_nov:.4f} pontos percentuais")

print("\nDetalhamento por produto:")
display(df[['Produto', 'Variação Nov/2024 (%)', 'Variação Dez/2024 (%)', 'Peso Médio']])
```
